# Cleaning columns names

```python
list(df)
```

```python
def proper_name(col):
    col = col.replace('>','')
    col = col.replace('<','')
    col = col.lower()
    return col

def proper_col(df):
    proper_df = pd.DataFrame()
    for col in list(df):
        proper_col = proper_name(col)
        proper_df[proper_col] = df[col]
    return proper_df
```

```python
proper_name(list(df)[0])
```

```python
proper_col(df)
```
