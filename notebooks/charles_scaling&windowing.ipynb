```python
from sklearn.preprocessing import StandardScaler

def std_scaler(df):
    """
    Scale the data with SKlearn StandardScaler
    """
    
    scaler = StandardScaler()
    
    col_to_scale_df = df.drop(columns=['ticker', 'date', '5TD_return', '10TD_return', '20TD_return'])
    col_to_scale = list(col_to_scale_df)
    
    scaled_df = df
    
    for col in col_to_scale:
        scaled_df[col] = scaler.fit_transform(scaled_df[[col]])
    
    return scaled_df
```

```python
scaled_df = std_scaler(df)
scaled_df.head()
```

```python
def windowing(df, strides=30):
    """
    Split the data into sets of n windows of size s==strides
    The data is ordered from the newest values to the oldest ones
    Checking if the last set is not lower than 30, skip it otherwise
    """
    
    working_df = df.copy()
    working_df = working_df.sort_values(by='date', ascending=False)
    
    windowed_data = []
    window = []
    
    o = 0
    #i = 0
    
    for i in range(df.shape[0]):
        if i >= df.shape[0]:
            break
        window.append(np.array(reversed_df.iloc[i,]))
        o += 1
        if o % strides == 0: # exclude automatically the last batch with not enough data
            windowed_data.append(window)
            window = []
            print(o)
        #i += 1

    return np.array(windowed_data)
```
